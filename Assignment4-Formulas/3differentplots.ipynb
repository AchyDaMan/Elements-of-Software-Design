```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Horizontal Axis = 5

```python
msize = 5
x = np.arange(0, msize, 0.1)
plt.plot(x, (2**20) * (x) + 2, 'red', x, x**7.1 + 2.1**20, 'blue', x, 4**x - 2.1**8, 'green')
plt.xlim(0, msize)
plt.rcParams["figure.figsize"] = (7,7)
plt.show()
```

# Horizontal Axis = 10

```python
msize = 10
x = np.arange(0, msize, 0.1)
plt.plot(x, (2**20) * (x) + 2, 'red', x, x**7.1 + 2.1**20, 'blue', x, 4**x - 2.1**8, 'green')
plt.xlim(0, msize)
plt.rcParams["figure.figsize"] = (7,7)
plt.show()
```

# Horizontal Axis = 15

```python
msize = 15
x = np.arange(0, msize, 0.1)
plt.plot(x, (2**20) * (x) + 2, 'red', x, x**7.1 + 2.1**20, 'blue', x, 4**x - 2.1**8, 'green')
plt.xlim(0, msize)
plt.rcParams["figure.figsize"] = (7,7)
plt.show()
```

# What do you see in the 3 plots?

It seems that when the true scale of the functions come out the farther the x-axis gets increased. You can see the linearity of the graph in both the x = 5 and x = 10 scales, but at x = 15 then it's not visible anymore.
